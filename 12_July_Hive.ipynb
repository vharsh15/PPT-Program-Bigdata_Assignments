```python
##1 
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Create the Employees table
create_table_query = """
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
"""

# Execute the CREATE TABLE query
with conn.cursor() as cursor:
    cursor.execute(create_table_query)

# Close the connection
conn.close()
```

```python
##2
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Define the query to retrieve records
query = "SELECT * FROM Customers WHERE age > 30"

# Execute the query and fetch the results
with conn.cursor() as cursor:
    cursor.execute(query)
    results = cursor.fetchall()

# Process the retrieved records
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
##3
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Define the query to sort records in descending order based on "timestamp" column
query = "SELECT * FROM Logs ORDER BY timestamp DESC"

# Execute the query and fetch the results
with conn.cursor() as cursor:
    cursor.execute(query)
    results = cursor.fetchall()

# Process the sorted records
for row in results:
    print(row)

# Close the connection
conn.close()
```

```python
##5
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Define the query to calculate the average salary
query = "SELECT AVG(salary) FROM Employees"

# Execute the query and fetch the result
with conn.cursor() as cursor:
    cursor.execute(query)
    result = cursor.fetchone()

# Process the result
average_salary = result[0]
print("Average Salary:", average_salary)

# Close the connection
conn.close()
```

```python
##6
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Define the query to create the partitioned table
create_table_query = """
    CREATE TABLE Sales_Data (
        id INT,
        product STRING,
        quantity INT,
        revenue FLOAT
    )
    PARTITIONED BY (year INT, month INT)
"""

# Execute the CREATE TABLE query
with conn.cursor() as cursor:
    cursor.execute(create_table_query)

# Define the query to add partitions to the table
add_partition_query = """
    ALTER TABLE Sales_Data ADD PARTITION (year=2022, month=1)
    LOCATION '/sales_data/year=2022/month=1'
"""

# Execute the ADD PARTITION query
with conn.cursor() as cursor:
    cursor.execute(add_partition_query)

# Close the connection
conn.close()
```

```python
##8
from pyhive import hive

# Establish a connection to Hive
conn = hive.connect(host='<hostname>', port=<port>, username='<username>')

# Define the query to perform the inner join
query = """
    SELECT o.order_id, o.order_date, c.customer_name
    FROM Orders o
    INNER JOIN Customers c ON o.customer_id = c.customer_id
"""

# Execute the query and fetch the results
with conn.cursor() as cursor:
    cursor.execute(query)
    results = cursor.fetchall()

# Process the retrieved records
for row in results:
    print(row)

# Close the connection
conn.close()
```
